## 10) Top 5 highest OBP (on base percentage) with at least 500 PAs in 1977?  (Dataframe)

```python
bbE = df
bbE['PA'] = bbE['AB'] + bbE['BB'] + bbE['IBB'] + bbE['SH'] + bbE['SF']
bbE['Avg'] = bbE['H'] / bbE['AB']
bbE['OBP'] = (bbE['H'] + bbE['BB'] + bbE['IBB'] + bbE['SH'] + bbE['SF'])/ bbE['AB']
```

```python
q10= bbE[bbE['yearID'] == 1977]
```

```python
q10n = q10.groupby('playerID').sum()
```

```python
q10nn = q10n[q10n['PA'] >= 500]
```

```python
q10nn.nlargest(5,'OBP')
```

## 12) Leaders in hits from 1940 up to and including 1949. (Dataframe)

```python
num_years = np.linspace(1940, 1949,10).astype(np.int64)
df3 = pd.read_csv('Batting.csv')
df_leaders = {}
count = 0
for i in num_years:
    y = df3[df3['yearID'] == i].groupby('playerID').sum().nlargest(1, 'H')
    df_leaders[count] = pd.DataFrame(y)
    count = count + 1
value = pd.concat(df_leaders)
value
```
